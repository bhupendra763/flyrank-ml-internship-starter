```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
overall_decline_rate = (df["GPA_Change"] < 0).mean()
decline_by_burnout = df.groupby("Burnout_Risk_Level")["GPA_Change"].apply(lambda g: (g < 0).mean())

print(f"Overall GPA-decline rate: {overall_decline_rate:.1%}")
print(decline_by_burnout.sort_values(ascending=False))
```

```python
# Number 2: does weekly GenAI usage actually differ across burnout risk groups?
genai_by_burnout = df.groupby("Burnout_Risk_Level")["Weekly_GenAI_Hours"].mean()
print(genai_by_burnout.sort_values(ascending=False))
```

```python
# Number 3: what correlates with GPA change at all? (sanity check the signal exists)
corr = df[[
    "Weekly_GenAI_Hours", "Traditional_Study_Hours",
    "Perceived_AI_Dependency", "Anxiety_Level_During_Exams",
    "Skill_Retention_Score", "GPA_Change"
]].corr()["GPA_Change"].sort_values(ascending=False)
print(corr)
```

### What these numbers say

- Overall, about **12.4%** of students finish the semester with a lower GPA than they started with. Among students already labeled **High Burnout Risk**, that rate increases to **14.8%**, compared with **11.3%** for the Medium group. This suggests there is a meaningful subgroup that may benefit from earlier support.

- Students in the **High Burnout Risk** group use GenAI for an average of **15.2 hours per week**, compared with **4.6 hours per week** for the Low Burnout Risk group—more than **three times higher**. This indicates that AI usage patterns may provide useful predictive information.

- Traditional study hours have the strongest positive relationship with GPA change (**correlation ≈ 0.38**), while GenAI usage hours and perceived AI dependency show only weak negative correlations (**≈ -0.04 to -0.05**). These are modest associations, so the project focuses on prediction rather than claiming that AI usage causes burnout.

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, "predicting Google").*

- **I can say:** In this dataset, high GenAI usage and high burnout risk tend to occur together, and students with higher burnout risk show a somewhat higher rate of GPA decline. These patterns can serve as an early-warning signal for decision support, but they do not explain why the relationship exists.

- **I cannot say:** GenAI use *causes* burnout or GPA decline. This dataset is observational, based on self-reported responses collected over a single semester. Students who are already struggling may simply rely on GenAI more, which could produce the same correlation.

- **I cannot say:** These findings automatically generalize to students at other universities, academic programs, or semesters. External validation on independent datasets would be required before making broader claims.

- The observed correlations (approximately **-0.04 to -0.05**) are relatively weak. Any predictive model should therefore be compared against a simple baseline, and its value should be judged by whether it provides a meaningful improvement in decision support rather than by claiming perfect or highly accurate predictions.
